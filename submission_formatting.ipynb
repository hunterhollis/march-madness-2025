```python
import pandas as pd
```

# Basic Predictions

```python
m_submissions = pd.read_csv('submissions/m_submission.csv')
print(len(m_submissions))
m_submissions.head()
```

```python
w_submissions = pd.read_csv('submissions/w_submission.csv')
print(len(w_submissions))
w_submissions.head()
```

```python
submissions = pd.concat([m_submissions, w_submissions]).reset_index(drop=True)
display(submissions)
submissions.to_csv('submissions/final_submission.csv', index=False)
```
